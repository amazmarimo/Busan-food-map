```python
import pandas as pd
from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
import numpy as np
 # seconds


# 원래 default는 caps["pageLoadStrategy"] = "normal"
```

```python
df = pd.read_csv('gudong.csv', sep=',')

# , encoding='CP949'
# '행정동명' 뺏음
df = df[['사업장명', '소재지전체주소','gu','dong','업태구분명', '소재지전화']] # 나중에 동으로 구분한 컬럼도 넣어야합니다.
df.columns = ['name', 'address','district','street' ,'cate1', 'phone']

 # 위와 같음
df = df.drop_duplicates(['name'], keep='first')

# df['cate_mix'] = df['cate1'] + df['cate2']

driver = webdriver.Chrome('chromedriver.exe') # seconds
driver.implicitly_wait(1)
df['kakao_keyword'] = '부산'+" "+df['district']+ " " + df['name']

search_list=[]

for i, keyword in enumerate(df['kakao_keyword'].tolist()):
    try:
        kakao_map_search_url = f"https://map.kakao.com/?q={keyword}"
        driver.get(kakao_map_search_url)
        
        rate = driver.find_element_by_css_selector("#info\.search\.place\.list > li:nth-child(1) > div.rating.clickArea > span.score > em").text
        rateNum = driver.find_element_by_css_selector("#info\.search\.place\.list > li:nth-child(1) > div.rating.clickArea > a > em").text
        url= driver.find_element_by_css_selector("#info\.search\.place\.list >li:nth-child(1)>div.rating.clickArea>a")
        url=url.get_attribute('href')
        search_list.append([keyword, rate, rateNum, url])

    except Exception as e1:
        print(str(i)+" 정보 없음")
        search_list.append([keyword,np.nan,np.nan,np.nan])
        pass
    
df_list = pd.DataFrame(search_list,columns=['사업장명', '평점', '리뷰','URL'])
```
